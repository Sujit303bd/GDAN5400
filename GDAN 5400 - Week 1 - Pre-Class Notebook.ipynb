# How to Share a Google Colab Notebook

## Steps to Share a Google Colab Notebook

1. **Open the Share Settings**:
   - Look at the top-right corner of your Colab interface.
   - Click the **Share** button (a button with a lock icon or a person icon).

2. **Adjust Sharing Permissions**:
   - In the **General access** dialog:
     - Under **Get link**, choose **Anyone with the link** if you want public access.
     - Choose one of the following roles:
       - **Viewer**: Others can only view your notebook.
       - **Commenter**: Others can add comments.
       - **Editor**: Others can edit your notebook.

3. **Copy and Share the Link**:
   - Copy the generated sharing link and submit that as your assignment submission on Canvas.
